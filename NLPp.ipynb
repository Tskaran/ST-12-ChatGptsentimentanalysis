```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("file.csv")
df.head()
```

```python
df.columns
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
df.columns
```

```python
# STEP-1: Data Cleaning
df["tweets"] = df["tweets"].str.casefold()
import re
mydata=[]
for i in df['tweets']:
  text = re.sub(r',','',i)
  cleaned_tweets = re.sub(r'[^\w\s]', '', text)
  cleaned_tweets = re.sub(r'\s+', ' ', cleaned_tweets)  
  cleaned_data = cleaned_tweets.strip()
  mydata.append(cleaned_data)
df['tweets']=mydata
df.head()
```

```python
print(df.head())
```

```python
# STEP-2: Tokenization
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
from nltk.tokenize import word_tokenize
```

```python
df.head()
```

```python
tokens=[word_tokenize(x) for x in df['tweets']]
```

```python
tokens[7]
```

```python
# STEP-3: Stopword Removal
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop=stopwords.words('english')
```

```python

stoptoken_list= []  # List to store stopword-removed tokens
for k in range(len(df['tweets'])):  # Loop through the tokenized text
    sw = [i for i in tokens[k] if i not in stop]  # Filter out tokens that are in the stopword list
    stoptoken_list.append(sw)
# stoptoken_list[7]
```

```python
# Step-4:pos_tagging
import re
import nltk
nltk.download('averaged_perceptron_tagger')
from nltk import pos_tag
```

```python
from nltk import pos_tag_sents

pos_token = pos_tag_sents(stoptoken_list)
```

```python
pos_token
```

```python
# Step-5: Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
pos_token[0][0]
```

```python
get_wordnet_pos(pos_token[0][1][1])
```

```python
lemmatizer.lemmatize(pos_token[0][0][0],get_wordnet_pos(pos_token[0][0][1]))
```

```python
lem_data = []
for x in range(len(pos_token)):
  lem = [lemmatizer.lemmatize(pos_token[x][i][0],get_wordnet_pos(pos_token[x][i][1])) for i in range(len(pos_token[x]))]
  lem_data.append(lem)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
stemedata =[]
for message in stoptoken_list:
  st = [ps.stem(word) for word in message]
  stemedata.append(st)
```

```python
stem_vec=[' '.join(message) for message in stemedata]
```

```python
#applying count vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
x=cv.fit_transform(stem_vec)
```
